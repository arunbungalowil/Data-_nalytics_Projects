# Billionaires Statistics Dataset 

##### Description
This dataset contains statistics on the world's billionaires, including information about their businesses, industries, and personal details. It provides insights into the wealth distribution, business sectors, and demographics of billionaires worldwide.

##### Key Features
1. **rank**: The ranking of the billionaire in terms of wealth.
2. **finalWorth**: The final net worth of the billionaire in U.S. dollars.
3. **category**: The category or industry in which the billionaire's business operates.
4. **personName**: The full name of the billionaire.
5. **age**: The age of the billionaire.
6. **country**: The country in which the billionaire resides.
7. **city**: The city in which the billionaire resides.
8. **source**: The source of the billionaire's wealth.
9. **industries**: The industries associated with the billionaire's business interests.
10. **countryOfCitizenship**: The country of citizenship of the billionaire.
11. **organization**: The name of the organization or company associated with the billionaire.
12. **selfMade**: Indicates whether the billionaire is self-made (True/False).
13. **status**: "D" represents self-made billionaires (Founders/Entrepreneurs) and "U" indicates inherited or unearned wealth.
14. **gender**: The gender of the billionaire.
15. **birthDate**: The birthdate of the billionaire.
16. **lastName**: The last name of the billionaire.
17. **firstName**: The first name of the billionaire.
18. **title**: The title or honorific of the billionaire.
19. **date**: The date of data collection.
20. **state**: The state in which the billionaire resides.
21. **residenceStateRegion**: The region or state of residence of the billionaire.
22. **birthYear**: The birth year of the billionaire.
23. **birthMonth**: The birth month of the billionaire.
24. **birthDay**: The birth day of the billionaire.
25. **cpi_country**: Consumer Price Index (CPI) for the billionaire's country.
26. **cpi_change_country**: CPI change for the billionaire's country.
27. **gdp_country**: Gross Domestic Product (GDP) for the billionaire's country.
28. **gross_tertiary_education_enrollment**: Enrollment in tertiary education in the billionaire's country.
29. **gross_primary_education_enrollment_country**: Enrollment in primary education in the billionaire's country.
30. **life_expectancy_country**: Life expectancy in the billionaire's country.
31. **tax_revenue_country_country**: Tax revenue in the billionaire's country.
32. **total_tax_rate_country**: Total tax rate in the billionaire's country.
33. **population_country**: Population of the billionaire's country.
34. **latitude_country**: Latitude coordinate of the billionaire's country.
35. **longitude_country**: Longitude coordinate of the billionaire's country

##### Dealing with Unnecessary Columns

From above information we can see that there are 14 columns which have float data type, 2 columns have integer data type and 18 columns have object(string) data type. After the careful observation on the dataset i decided to drop some columns which are not relavent in this analysis. also `country` and `countryOfCitizenship` columns hold the same value so i will just keep one column.
following are the columns that i am going to drop.
- `city`: This column contains the city in which the billionaire resides. rather than `city` column i am interested in the citizenship of the billionaires.


- `category`: This column contains the same information of `industry` column


- `lastName`, `firstName`: There is a column called `personName`, which contains both first name and last name, so i decided to keep `personName` column.


- `countryOfCitizenship`: This column contains the same information of `country` column, so just keep `country` column.


- `date`: This column contains the date of data collection and it is not necessary for the analysis.


- `title`: The title or honorific of the billionaire


- `residenceStateRegion`, `state`: These columns may not be relevant for the analysis or may provide redundant information if there are other columns indicating the geographical location.


- `cpi_country`, `cpi_change_country`: These columns may contain information related to consumer price index (CPI) specific to a country, which may not be necessary depending on the analysis.


- `gross_tertiary_education_enrollment`, `gross_primary_education_enrollment_country`: These columns may represent education enrollment statistics that are not relevant for the analysis.

```python
data.drop(['city', 'category', 'lastName', 'firstName', 'countryOfCitizenship', 'date', 'title', 'state',
          'residenceStateRegion', 'cpi_country', 'cpi_change_country', 
          'gross_tertiary_education_enrollment', 'gross_primary_education_enrollment_country'],
          axis = 1, inplace = True)
```

```python
print(f' Total number of columns after dropping the irrelavant columns: {len(data.columns)}')
```

Next step is renaming the columns.

```python
data.rename(columns = {'finalWorth':'final_worth', 'personName':'person_name', 'selfMade': 'self_made',
                        'birthDate': 'birth_date', 'birthYear': 'birth_year','birthMonth':'birth_month',
                        'birthDay':'birth_day', 'gdp_country': 'gdp_country($)'}, inplace = True)
data.columns
```

- `age` column has float data type, it has to be converted to integer data type.


- `birth_date` column has object data type so i will convert that innto datetime.


- `'birth_year', 'birth_month', 'birth_day'` columns have float data type, they have to be converted to integer data type.


- `gdp_country` column has object data type because the values starts with $, so i will drop it and conver the column into 
float type.

```python
for column in data.columns:
    if column == 'age':
        data['age'] = data['age'].astype('Int64')
    elif column == 'birth_date':
        data['birth_date'] = pd.to_datetime(data['birth_date'])
    elif column == 'birth_year':
        data['birth_year'] = data['birth_year'].astype('Int64')
    elif column == 'birth_month':
        data['birth_month'] = data['birth_month'].astype('Int64')
    elif column == 'birth_day':
        data['birth_day'] = data['birth_day'].astype('Int64')
    elif column == 'gdp_country($)':
        data['gdp_country($)'] = data['gdp_country($)'].str.replace('$', '').replace(',', '', regex = True).astype(float)
```

##### Handling Null Values

```python
# total number of null values
print(f'Total number of null values in the dataset: {data.isnull().sum().sum()}') 
```

```python
def null_values(data):
    missing_values = data.isnull().sum()
    missing_percentage = (missing_values / len(data)) * 100
    result = pd.DataFrame({'number_of_missing_value':missing_values,
                           'percentage_of_missing_values(%)':missing_percentage.map('{:.2f}'.format)})
    return result
  
null_values(data)
```

Having initially streamlined my dataset by dropping unnecessary columns, my focus shifted to the crucial task of inspecting and managing null values. This stage is pivotal, as it involves a meticulous examination to identify any gaps or missing information in the dataset.

Utilizing descriptive statistics I carefully assessed the presence of null values in each column. This insightful analysis revealed potential data gaps and allowed me to quantify the extent of missing information. I documented both the count and percentage of null values in each column, providing a clear overview of the dataset's completeness.

To address these null values, I employed a thoughtful strategy tailored to the unique characteristics of my data. I considered techniques such as imputation, where missing values are replaced with statistical measures like the mean. Additionally, I evaluated the option of removing rows or columns when appropriate, ensuring that my dataset remains robust and suitable for further analysis.

After inspecting the data it is found that `organization` column has higher percentage of missing values `87.69%`, i checked the column and i decided not to keep the column for further analysis.

```python
data.drop('organization', axis = 1, inplace = True)
```

`country` column is object type i decided to replace the null values with `other`. for integer and float type columns i am going to take mean value and apply on each null values for this i have to define a function, which take the  datframe, column name and the value to be filled.

for the other columns, that we are going to use for `EDA`, i will use mean or median to fill the null values

```python
def replacing_missing_values(data, column, value):
    data[column].fillna(value, inplace = True)
replacing_missing_values(data, 'country', 'other')
```

```python
# replacing the remaing null values with mean value of the columns
columns_to_fill = ['longitude_country', 'latitude_country', 'population_country', 'total_tax_rate_country',
                  'tax_revenue_country_country','life_expectancy_country','gdp_country($)'] 

for column in columns_to_fill:
    mean_value = data[column].mean()
    data[column].fillna(mean_value, inplace = True)
```

#### Distribution of Male and Female Billionaires
In the initial stages of our Exploratory Data Analysis (EDA),i delved into the demographic distribution of billionaires in our dataset. One key facet of this exploration was the comprehensive tally of male and female billionaires.

```python
male_billionaires = len(data[data['gender'] == 'M'])
female_billionaires = len(data[data['gender'] == 'F'])
print(f'Total number of male billionaires: {male_billionaires}')
print(f'Total number of female billionaires: {female_billionaires}')
```

```python
genders = ['Male', 'Female']
counts = [male_billionaires, female_billionaires]
# This line creates a Matplotlib figure and axis. fig is the entire figure, and ax is the subplot or axis within that figure.
fig, ax = plt.subplots(figsize = (10,6)) 
bars = ax.bar(genders, counts, color = ['blue', 'pink']) # This line creates a bar plot using the bar function
for bar in bars: # iterate each bar in the plot
    yval = bar.get_height() # retrieve the height of the current bar
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval), ha = 'center', va = 'bottom') # add text annotation over each bar
plt.title('Distribution of Male and Female Billionaires')
plt.xlabel('Gender')
plt.ylabel('Number of Billionaires')
plt.show()
```

The dataset reveals a notable presence of both male and female billionaires, with a striking total of `2,303` individuals classified as male billionaires and `337` as female billionaires. ####  selfmade vs inherited

```python
# selfmade vs inherited
inherited = 0
selfmade = 0
for item in data['self_made']:
    if item == True:
        selfmade += 1
    else:
        inherited += 1
print(f'Number of self made billionaires: {selfmade}')
print(f'Number of inherited billionaires: {inherited}')
```

```python
title = ['self_made', 'inherited']
values = [selfmade, inherited]
fig, ax = plt.subplots(figsize = (10,6)) 
bars = ax.bar(title, values, color = ['blue', 'pink']) # This line creates a bar plot using the bar function
for bar in bars: # iterate each bar in the plot
    yval = bar.get_height() # retrieve the height of the current bar
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval), ha = 'center', va = 'bottom') # add text annotation over each bar
plt.title('Distribution of Inherited vs Self made Billionaires')
plt.xlabel('Inherited vs Self made')
plt.ylabel('Number of Billionaires')
plt.show()
```


In our comprehensive analysis of the billionaire landscape, a discerning distinction emerges between individuals who have forged their paths to extreme wealth and those who have inherited their financial legacies. The dataset reveals a formidable cohort of 1,812 self-made billionaires, representing individuals who, through their entrepreneurial acumen and innovation, have navigated the complex terrain of business and amassed extraordinary wealth.

Conversely, our analysis also sheds light on 828 individuals who have inherited their wealth, comprising a distinct group of billionaires whose fortunes have been passed down through familial lines. These inherited billionaires, while contributing to the overall wealth distribution, underscore the enduring impact of family legacies on the global billionaire ecosystem.

#### Age Group

```python
above_60 = 0
below_30 = 0
age_30_60 = 0
age_info = data.dropna(subset = ['age'])

for age in age_info['age'].values:
    if age > 60:
        above_60 += 1
    elif 30 <= age <= 60:
        age_30_60 += 1
    elif age < 30:
        below_30 += 1

age_group = [above_60, age_30_60, below_30]
titles = ['> 60', '30-60', '< 30']
fig, ax = plt.subplots(figsize = (10,6)) 
bars = ax.bar(titles, age_group, color = ['red', 'yellow','blue']) # This line creates a bar plot using the bar function
for bar in bars: # iterate each bar in the plot
    yval = bar.get_height() # retrieve the height of the current bar
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval), ha = 'center', va = 'bottom') # add text annotation over each bar
plt.title('Distribution of Billionaires by Age Group')
plt.show()
```

In our meticulous analysis of billionaires' age demographics, a fascinating narrative unfolds, revealing distinctive patterns across various age groups. The majority of billionaires, a substantial 1,572 individuals, fall into the category of seasoned leaders aged above 60. This group signifies a wealth of experience and a long-standing presence within the realm of extreme wealth, contributing significantly to the global economic landscape.

Contrastingly, a smaller cohort of 11 billionaires belongs to the age group below 30, showcasing a remarkable trend of youthful entrepreneurs who have rapidly ascended to extreme wealth at an early stage in their careers. This group represents a dynamic force driving innovation and disruption in various industries.

Between the two extremes, the age group of 30 to 60 years encompasses a robust 992 billionaires. This segment signifies a diverse range of individuals at the peak of their professional careers, leveraging their expertise and industry knowledge to amass substantial wealth.
