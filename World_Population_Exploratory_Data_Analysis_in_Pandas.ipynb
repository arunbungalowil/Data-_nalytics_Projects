## Introduction:
The World Population Dataset is a comprehensive collection of data that provides insights into global population trends and demographics over time. This dataset is a valuable resource for conducting data exploration and analysis, with a specific focus on understanding population growth trends across continents.

## Research Question:
The primary research question for this dataset exploration is: "What are the population growth trends per continent?"

### Columns in the Dataset:

1. Rank: The ranking of countries by population.
2. CCA3: The Country Code (e.g., ISO 3166-1 alpha-3 code).
3. Country: The name of the country or region.
4. Capital: The capital city of the country.
5. Continent: The continent to which the country belongs.
6. 2022 Population: The population count for the year 2022.
7. 2020 Population: The population count for the year 2020.
8. 2015 Population: The population count for the year 2015.
9. 2010 Population: The population count for the year 2010.
10. 2000 Population: The population count for the year 2000.
11. 1990 Population: The population count for the year 1990.
12. 1980 Population: The population count for the year 1980.
13. 1970 Population: The population count for the year 1970.
14. Area (km²): The land area of the country in square kilometers.
15. Density (per km²): The population density per square kilometer of land area.
16. Growth Rate: The population growth rate.
17. World Population Percentage: The percentage of the world's population represented by the country.

```python
df2 = df_data.groupby("Continent").mean().sort_values(by = "World Population Percentage", ascending = False)
df2
```

```python
df2 = df2[[
       '1970 Population', '1980 Population', '1990 Population',
       '2000 Population', '2010 Population', '2015 Population',
       '2020 Population','2022 Population']]

df2
```

```python
df3 = df2.transpose()
df3
```

## Conclusion:
From the graph illustrating population growth trends across continents, it is evident that there are distinct patterns in population dynamics. Notably, the data clearly demonstrates that the Asian continent has been experiencing significant population growth over the years. This growth trend suggests that Asia continues to be a region with a rapidly expanding population.

Conversely, the Oceania region stands out for its relatively stable population growth. While some fluctuations may occur, Oceania has maintained a more consistent growth rate compared to other continents.
