# Exploratory Data Analysis

The Roller Coaster dataset is a comprehensive collection of information on 1,087 roller coasters, represented by 56 columns.

#### Dataset Overview:
Size: The dataset consists of 1,087 rows and 56 columns.
Columns: The dataset includes information on coaster characteristics, such as name, length, speed, location, manufacturer, height, inversions, launch system, cost, and more.

#### Key Columns for EDA:

1. **Coaster Name**: The name of the roller coaster.
2. **Length**: The length of the roller coaster's track.
3. **Speed**: The maximum speed of the roller coaster.
4. **Location**: The location (park and city) where the coaster is situated.
5. **Status**: The operational status of the coaster.
6. **Opening Date**: The date when the coaster was opened to the public.
7. **Type**: The type or category of the coaster.
8. **Manufacturer**: The company that manufactured the roller coaster.
9. **Height**: The height of the coaster.
10. **Inversions**: The number of inversions the coaster has.
11. **Cost**: The estimated cost to build the roller coaster.

## Data Preparation

In the data preparation phase, i am going to shape the dataset to focus on specific columns that are relevant to the  analysis. This process often involves selecting, renaming, and cleaning the data to ensure it is in a suitable format for the exploration.
In focusing on these specific columns, I'm narrowing down the dataset to essential features that are highly relevant to my analysis goals. By removing unnecessary columns, I've made the data more manageable and conducive to uncovering meaningful insights about roller coasters, their locations, and key characteristics. This streamlined dataset now serves as the foundation for my subsequent data analysis and visualization tasks.

```python
df = df[['coaster_name', 
    #'Length', 'Speed', 
    'Location', 'Status', 
    #'Opening date',
    #'Type', 
    'Manufacturer', 
    #'Height restriction', 'Model', 'Height',
    #'Inversions', 'Lift/launch system', 'Cost', 'Trains', 'Park section',
    #'Duration', 'Capacity', 'G-force', 'Designer', 'Max vertical angle',
    #'Drop', 'Soft opening date', 'Fast Lane available', 'Replaced',
    #'Track layout', 'Fastrack available', 'Soft opening date.1',
    #'Closing date',
    #'Opened',
    #'Replaced by', 'Website',
    #'Flash Pass Available', 'Must transfer from wheelchair', 'Theme',
    #'Single rider line available', 'Restraint Style',
    #'Flash Pass available', 'Acceleration', 'Restraints', 'Name',
    'year_introduced',
    'latitude', 
    'longitude',
    'Type_Main',
    'opening_date_clean',
    #'speed1', 'speed2', 'speed1_value', 'speed1_unit',
     'speed_mph', 
    #'height_value', 'height_unit', 
    'height_ft',
    'Inversions_clean', 'Gforce_clean']].copy()
```

In the process of checking for null values within the DataFrame, I observed that various columns exhibit different degrees of missing data. Understanding the distribution of missing values in the dataset is crucial for data cleaning and analysis.
For the column *Height_ft* there are 916 missing values, i have total 1087 entries, so it would be better to drop the column,
to tackle the missing value problem, i will remove the rows with missing values.

```python
df.drop('Height_ft', axis = 1, inplace = True)
```

```python
df
```

```python
df.dropna(subset = ["Status", "Opening Date"], inplace = True)
```

```python
df
```

Following the removal of rows with missing values in the "Status" and "Opening Date" columns, the Roller Coaster dataset now consists of 828 rows. Notably, rows with missing data in these two columns were selectively eliminated, leaving a more streamlined dataset for analysis.

However, rows with missing values in the "Manufacturer," "Latitude," "Longitude," "Speed mph," and "Gforce" columns have been retained. those empty cells for the numerical columns are filled with the mean values.

```python
columns_to_fill = ["Gforce", "Speed mph", "Latitude", "Longitude"]
for column in columns_to_fill:
    df[column].fillna(df[column].mean(), inplace =True)
```

```python
df.isna().sum()
# df
```

The next stage in my data preparation process involves checking for duplicate entries within the dataset and subsequently removing those duplicates. This step is essential for ensuring data integrity and accuracy in my analysis. Duplicate entries can lead to distorted results and affect the reliability of my findings. By identifying and eliminating duplicates, I streamline my dataset, resulting in cleaner, more trustworthy data that serves as a solid foundation for my subsequent analysis and modeling. This de-duplication process is pivotal for maintaining the quality and reliability of the data, which, in turn, enhances the robustness of my data-driven insights and conclusions.

```python
df.loc[df.duplicated(subset = ['Coaster Name', 'Location','Opening Date'])]
```

```python
df = df.loc[~df.duplicated(subset = ['Coaster Name', 'Location','Opening Date'])].reset_index(drop = True)
```

## Feature understanding

For my initial visualization, I'm embarking on a journey through the world of roller coasters by crafting an engaging bar plot that unveils the top ten roller coasters based on their year of introduction. 

```python
plt.figure(figsize = (10, 5))
ax = df['Year Introduced'].value_counts()\
                     .head(10)\
                     .plot(kind = 'bar', color = 'blue', title = 'Roller coaster intoduction year')
ax.set_xlabel('Year introduced')
ax.set_ylabel('Total number')
plt.show()
```

```python
df['Speed mph'].plot(kind = 'hist', bins = 20)
plt.show()
```

A histogram for the "Speed mph" column is a visual representation that provides insights into the distribution of roller coaster speeds in miles per hour, A histogram for the "Speed mph" column that predominantly falls within the range of 30 to 60 miles per hour suggests a common speed interval for roller coasters in the dataset. This concentration of data within this specific range indicates that many roller coasters operate at speeds between 30 and 60 mph.

```python
top_locations = df["Location"].value_counts().head(10)
top_locations = top_locations[top_locations.index != 'Other']
top_locations.plot(kind = 'pie', title = 'Top Locations')
plt.show()
```

Explore correlations between coaster features (e.g., speed) and year of introduction

```python
plt.figure(figsize = (10, 5))
correlation = df[["Year Introduced", "Speed mph"]].corr()
sns.heatmap(correlation, annot = True)
plt.title('Correlation between speed and year of introduction')
plt.show()
```
