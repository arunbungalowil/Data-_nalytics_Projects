## Euro Daily Exchange Rate
It wasn't until 1999 that the euro really began its journey, when 11 countries (Austria, Belgium, Finland, France, Germany, Ireland, Italy, Luxembourg, the Netherlands, Portugal and Spain) fixed their exchange rates and created a new currency with monetary policy passed to the European Central Bank. Today euro is 20+ years old.

### Content
Reference rates are euro foreign exchange rates observed on major foreign exchange trading venues at a certain point in time they are the price of one currency in terms of another currency. The rates are usually updated around 16:00 CET on every working day, except on TARGET closing days.

Dataset contains date and Euro rate corresponding to Australian dollar, Bulgarian lev, Brazilian real, Canadian dollar, Swiss franc, Chinese yuan renminbi, Cypriot pound, Czech koruna, Danish krone, Estonian kroon, UK pound sterling, Greek drachma, Hong Kong dollar, Croatian kuna, Hungarian forint, Indonesian rupiah, Israeli shekel, Indian rupee, Iceland krona, Japanese yen, Korean won, Lithuanian litas, Latvian lats, Maltese lira, Mexican peso, Malaysian ringgit, Norwegian krone, New Zealand dollar, Philippine peso, Polish zloty, Romanian leu, Russian rouble, Swedish krona, Singapore dollar, Slovenian tolar, Slovak koruna, Thai baht, Turkish lira, US dollar, South African rand.

Some currency in the list doesn't exist anymore; it was replaced by the Euro â‚¬: `Cypriot pound (2007)`, `Estonian kroon (2011)`, `Greek drachma (2002)`, `Lithuanian litas (2015)`, `Latvian lats (2014)`, `Maltese lira (2008)`, `Slovenian tolar (2007)`, `Slovak koruna (2009)`.


```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import matplotlib.style as style 
import seaborn as sns
pd.set_option('display.max_columns', 100)
style.use('ggplot')
%matplotlib inline
```

```python
exchange_rates = pd.read_csv('data/euro-daily-hist_1999_2023.csv')
exchange_rates.head(5)
```

```python
# checking the column names and datatypes
exchange_rates.info()
```

```python
exchange_rates.head()
```

### Cleaning the Column Names

```python
def clean_column_names(column_name):
    column = (column_name.replace('[', '')
                         .replace(']', '')
                         .strip()
                         .replace(' ','_')
                         .replace('\\', '_'))
    if column == 'Period_Unit:':
        column = 'Time'
    return column         
```

```python
updated_column_names = []
for column in exchange_rates.columns:
    updated_column_names.append(clean_column_names(column))
print(updated_column_names)
```

```python
exchange_rates.columns = updated_column_names
```

### Transforming Time Data: Converting `Time` Column to Date Format

```python
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time', inplace = True) # sort the dataset in ascending order of date column
exchange_rates.reset_index(drop = True, inplace = True) # reset index and drop the old index
exchange_rates.head()
```

### Streamlining Currency Data: Dropping Obsolete Currencies

To ensure our dataset remains relevant and focused, we're removing columns representing currencies that are no longer in use. By dropping the following obsolete currencies, we streamline our data for better clarity:

1. Cypriot_pound
2. Estonian_kroon  
3. Greek_drachma
4. Lithuanian_litas 
5. Latvian_lats 
6. Maltese_lira 
7. Slovenian_tolar  
8. Slovak_koruna 

```python
exchange_rates.drop(['Cypriot_pound','Estonian_kroon','Greek_drachma','Lithuanian_litas',
                     'Latvian_lats','Maltese_lira','Slovenian_tolar','Slovak_koruna'], axis =1, inplace = True)
```

### Exploratory data visualisation
#### Euro to Dollar exchange rates

We've extracted a subset of our dataset focusing on the Euro to Dollar exchange rates, containing two key columns: `Time` and `US_dollar`. This curated data snapshot provides a clear view of the temporal dynamics of the Euro against the US Dollar. By examining this subset, we can uncover trends, fluctuations, and patterns in the exchange rates over time, facilitating informed decision-making and deeper insights into currency market behavior.

```python
euro_to_dollar = exchange_rates[['Time', 'US_dollar']].copy()
euro_to_dollar.head()
```

```python
euro_to_dollar['US_dollar'].value_counts()
```

By removing rows where the 'US_dollar' value is represented as '-', we ensure that our analysis is based on valid numerical data points. 

```python
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype('float')
euro_to_dollar.info()
```

```python
# Euro to Dollar rate
locations = ['left', 'right', 'top', 'bottom']
fig, ax = plt.subplots(figsize = (12, 6))
plt.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar'], color = 'b')
for location in locations:
    ax.spines[location].set_visible(False)
ax.tick_params(bottom = False, left =False)
plt.show()
```

If we look at the line's shape, we see many small wiggles â€” rather than seeing a smooth line. The wiggles, however, have meaning: : they are the visual representation of the daily variation in the exchange rate. The rate goes up and down, up and down again, day to day. The rate only shows clear upward or downward trends in the longer run (months or years).

```python
euro_to_dollar.set_index('Time', inplace=True)
weekly_data = euro_to_dollar['US_dollar'].resample('W').mean()
monthly_data = euro_to_dollar['US_dollar'].resample('M').mean()
yearly_data = euro_to_dollar['US_dollar'].resample('Y').mean()
```

```python
print(weekly_data[:5])
print('-----------------')
print(monthly_data[:5])
print('-----------------')
print(yearly_data[:5])
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(3,1, figsize = (12,15))
axeses = (ax1, ax2, ax3)
ax1.plot(weekly_data.index, weekly_data, color = 'b')
ax1.set_title('Weekly Exchange Rate')
ax1.set_ylabel('Exchange Rate (USD)')
for location in locations:
    ax1.spines[location].set_visible(False)
ax1.tick_params(bottom = False, left =False)
ax1.set_xticklabels([])

ax2.plot(monthly_data.index, monthly_data, color = 'b')
ax2.set_title('Monthly Exchange Rate')
ax2.set_ylabel('Exchange Rate (USD)')
for location in locations:
    ax2.spines[location].set_visible(False)
ax2.tick_params(bottom = False, left =False)
ax2.set_xticklabels([])

ax3.plot(yearly_data.index, yearly_data, marker = 'o', color = 'b')
ax3.set_title('yearly Exchange Rate')
ax3.set_xlabel('Date')
ax3.set_ylabel('Exchange Rate (USD)')
for location in locations:
    ax3.spines[location].set_visible(False)
ax3.tick_params(bottom = False, left =False)


plt.show()
```


### Here are a few story ideas for our data:
